```python
  import matplotlib.pyplot as plt
  fig, axs = plt.subplots(1, 2, figsize=(12, 7))
  berg = Iceberg(length=733, dz=5)
  berg.plot_iceberg_shape(ice=berg.init_iceberg_size(), ax=axs[0])          # Barker
  berg.plot_iceberg_shape(ice=berg.init_iceberg_size(
      keel_method='schild', volume_law='sulak'), ax=axs[1])                 # calibrated
  axs[0].set_title('Uncalibrated (Barker)'); axs[1].set_title('Calibrated (Schild/Sulak)')
```
